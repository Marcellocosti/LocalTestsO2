```python
import uproot
import pandas as pd
import matplotlib.pyplot as plt
import awkward as ak

tables = [
    "O2hfredcollision",
    "O2hfredcollcent",
    # "O2hfredqvector",     # NOT USED
    "O2hfredcollextra",
    "O2hforigcolcount",
    "O2hfredtrackbase",
    "O2hfredtrackcov",
    "O2hfredtrackpid",
    "O2hfredtrackmom",
    "O2hfred2prong",
    "O2hfred2prongcov",
    "O2hfred3prongml_001",
    "O2hfredmomddaugs",
    "O2hfredsoftpibase",
    "O2hfredsoftpicov",
    "O2hfredsoftpipid",
    "O2hfredpiddau0_001",
    "O2hfredpiddau1_001",
    "O2hfcandb0config",
    "O2hfseldstar"
    # "O2hfmcgenredb0",    # NOT USED
    # "O2hfmcrecreddstpi"  # NOT USED
]

with uproot.open("Tree.root") as f:
    keys = f.keys()
    print("ROOT Keys:", keys)
    
    for table in tables:
        # Filter keys that match the table name
        matched_keys = [key for key in keys if table in key]
        
        if not matched_keys:
            print(f"No keys found for table: {table}")
            continue
        
        dfs = []
        for key in matched_keys:
            print(f"Reading {key}")
            array = f[key].arrays(library="ak")  # get awkward array
            df = ak.to_dataframe(array)          # convert to pandas DataFrame
            dfs.append(df)
        
        full_df = pd.concat(dfs, ignore_index=True)
        print(f"\nTable: {table}")
        print(f"Number of entries: {len(full_df)}")
        print(f"Columns: {full_df.columns.tolist()}")
        
        # Plotting
        n_columns = len(full_df.columns)
        n_rows = (n_columns // 4) + (1 if n_columns % 4 else 0)
        
        fig, axes = plt.subplots(n_rows, 4, figsize=(15, 5 * n_rows))
        axes = axes.flatten()

        for i, column in enumerate(full_df.columns):
            print(f"Plotting {column}")
            try:
                axes[i].hist(full_df[column].dropna(), bins=50, color='blue')
                axes[i].set_title(f'{column}')
                axes[i].set_xlabel('Value')
                axes[i].set_ylabel('Frequency')
            except Exception as e:
                print(f"Could not plot column {column}: {e}")
                axes[i].set_visible(False)
        
        # Remove unused axes if any
        for i in range(n_columns, len(axes)):
            fig.delaxes(axes[i])
        
        fig.suptitle(f"Histograms for table: {table}", fontsize=16)
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()
```

```python
import uproot
import awkward as ak
import matplotlib.pyplot as plt

tables = ["O2hfmldstar"]

with uproot.open("Tree.root") as f:
    keys = f.keys()
    print("ROOT Keys:", keys)

    for table in tables:
        matched_keys = [key for key in keys if table in key]
        if not matched_keys:
            print(f"No keys found for table: {table}")
            continue

        for key in matched_keys:
            print(f"Reading {key}")
            array = f[key].arrays(library="ak")

            # Pick the vector column
            if "fMlProbDstarToD0Pi" not in array.fields:
                print("No fMlProbDstarToD0Pi in this tree")
                continue

            vecs = array["fMlProbDstarToD0Pi"]
            print("Column type:", vecs.type)

            # Determine maximum number of components
            n_components = ak.max(ak.num(vecs))
            print(f"Vector has up to {n_components} components")

            # Plot each position separately
            for i in range(n_components):
                # Extract i-th component across all events (if present)
                comp = vecs[ak.num(vecs) > i][:, i]
                plt.figure()
                plt.hist(ak.to_numpy(comp), bins=50, color="blue", alpha=0.7)
                plt.title(f"fMlProbDstarToD0Pi[{i}]")
                plt.xlabel("Value")
                plt.ylabel("Frequency")
                plt.show()
```
